```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample input data
input_data = (62, 0, 0, 140, 268, 0, 0, 160, 0, 3.6, 0, 2, 2)

# Change the input data to a numpy array and reshape it
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Make prediction
prediction = model.predict(input_data_reshaped)

# Determine prediction label
prediction_label = 'The Person does not have a Heart Disease' if prediction[0] == 0 else 'The Person has Heart Disease'

# Plotting the prediction
plt.figure(figsize=(8, 6))
plt.boxplot(input_data)
plt.title('Prediction Result')
plt.xlabel('Features')
plt.ylabel('Values')
plt.text(1, max(input_data) - 10, prediction_label, fontsize=12, bbox=dict(facecolor='green', alpha=0.5))
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
heart_data = pd.read_csv('D:\capstone final project\multiple-disease-prediction-streamlit-app-main\dataset\heart.csv')

# Prepare the data
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']

# Split the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

# Initialize Logistic Regression model
model = LogisticRegression()
model.fit(X_train, Y_train)

# Calculate accuracy scores
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy)

# Generate Seaborn bar plot for accuracy
data = {
    'Data Type': ['Training Data', 'Test Data'],
    'Accuracy Score': [training_data_accuracy, test_data_accuracy]
}

sns.set(style="whitegrid")
ax = sns.barplot(x="Data Type", y="Accuracy Score", data=data)

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.title('Accuracy Scores of Training and Test Data for Logistic Regression')
plt.show()

# Generate classification report for test data
report = classification_report(Y_test, X_test_prediction, target_names=['No Heart Disease', 'Heart Disease'])
print("Classification Report for Test Data:\n", report)

# Plot classification report
plt.figure(figsize=(8, 6))
sns.heatmap(pd.DataFrame(classification_report(Y_test, X_test_prediction, output_dict=True)).T, annot=True)
plt.title('Classification Report for Test Data')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load the dataset
heart_data = pd.read_csv('D:\capstone final project\multiple-disease-prediction-streamlit-app-main\dataset\heart.csv')

# Prepare the data
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']

# Split the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

# Initialize Logistic Regression model
model = LogisticRegression()
model.fit(X_train, Y_train)

# Calculate accuracy scores
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy)

# Create DataFrame for pair plot
pairplot_df = X_train.copy()
pairplot_df['Actual'] = Y_train
pairplot_df['Predicted'] = X_train_prediction

# Set style
sns.set(style="ticks")

# Generate pair plot
sns.pairplot(pairplot_df, hue="Actual", markers=["o", "s"], palette="husl", diag_kind='kde', diag_kws=dict(shade=True))

# Title for the plot
plt.suptitle('Pair Plot for Logistic Regression Accuracy and Test/Train Data', y=1.02)

# Show the plot
plt.show()
```
